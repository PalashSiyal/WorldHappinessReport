```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from matplotlib import style
```

# **Introduction**
**Visualizing and Comparing the Ladder score, GDP.**

```python
df=pd.read_csv("world-happiness-report-2021.csv")
df.head(5)
```

```python
country_name_dict = {"Luxembourg*": "Luxembourg", 
                    "Guatemala*": "Guatemala",
                    "Kuwait*": "Kuwait",
                    "Belarus*": "Belarus",
                    "Turkmenistan*": "Turkmenistan",
                    "North Cyprus*": "North Cyprus",
                    "Libya*": "Libya",
                    "Azerbaijan*": "Azerbaijan",
                    "Gambia*": "Gambia",
                    "Liberia*": "Liberia",
                    "Niger*": "Niger",
                    "Comoros*": "Comoros",
                    "Palestinian Territories*": "Palestinian Territories",
                    "Madagascar*": "Madagascar",
                    "Chad*": "Chad",
                    "Yemen*": "Yemen",
                    "Mauritania*": "Mauritania",
                    "Lesotho*": "Lesotho",
                    "Botswana*": "Botswana",
                    "Rwanda*": "Rwanda",}
df["Country name"] = df["Country name"].replace(country_name_dict)
df.head()
```

```python
continents_df = pd.read_csv("world-happiness-report-2021.csv")
continents_dict = {k:v for k,v in zip(continents_df["Country name"],continents_df["Regional indicator"])}
df["Continents"] = df["Country name"].map(continents_dict)
```

**Countries**

```python
#asian countries
Southeast_Asia=df[df['Continents']== 'Southeast Asia']
South_Asia=df[df['Continents']== 'South Asia']
East_Asia=df[df['Continents']== 'East Asia']
```

```python
#european countries
Western_Europe=df[df['Continents']== 'Western Europe']
Central_and_Eastern_Europe=df[df['Continents']== 'Central and Eastern Europe']
```

```python
#middle east and african countries
Middle_East_and_North_Africa=df[df['Continents']== 'Middle East and North Africa']
Sub_Saharan_Africa=df[df['Continents']== 'Sub-Saharan Africa']
```

```python
#american countries
North_America_and_ANZ=df[df['Continents']== 'North America and ANZ']
Latin_America_and_Caribbean=df[df['Continents']== 'Latin America and Caribbean']
```

```python
#independent states
Commonwealth_of_Independent_States=df[df['Continents']== 'Commonwealth of Independent States']
```

# **Ladder score**

```python
#WORLD
plt.figure(figsize=(35,10))
sns.barplot(data=df,x="Country name",y="Ladder score")
plt.xticks(rotation=90) 
plt.show()
```

```python
plt.figure(figsize = (16,8))
plt.xticks(rotation=90)
sns.lineplot(x="Country name", y = "Ladder score", data = Southeast_Asia,marker='o',label='Southeast_Asia')
sns.lineplot(x="Country name", y = "Ladder score", data = South_Asia,marker='o',label='South_Asia')
sns.lineplot(x="Country name", y = "Ladder score", data = East_Asia,marker='o',label='East_Asia')
plt.show()
```

```python
plt.figure(figsize = (16,8))
plt.xticks(rotation=90)
sns.lineplot(x="Country name", y = "Ladder score", data = Western_Europe,marker='o',label='Western_Europe')
sns.lineplot(x="Country name", y = "Ladder score", data = Central_and_Eastern_Europe,marker='o',label='Central_and_Eastern_Europe')
plt.show()
```

```python
plt.figure(figsize = (16,8))
plt.xticks(rotation=90)
sns.lineplot(x="Country name", y = "Ladder score", data =Middle_East_and_North_Africa,marker='o',label='Middle_East_and_North_Africa')
sns.lineplot(x="Country name", y = "Ladder score", data = Sub_Saharan_Africa,marker='o',label='Sub_Saharan_Africa')
plt.show()
```

```python
plt.figure(figsize = (16,8))
plt.xticks(rotation=90)
sns.lineplot(x="Country name", y = "Ladder score", data =North_America_and_ANZ,marker='o',label='North_America_and_ANZ')
sns.lineplot(x="Country name", y = "Ladder score", data =Latin_America_and_Caribbean,marker='o',label='Latin_America_and_Caribbean')
plt.show()
```

```python
plt.figure(figsize = (16,8))
plt.xticks(rotation=90)
sns.lineplot(x="Country name", y = "Ladder score", data =Commonwealth_of_Independent_States,marker='o',label='Commonwealth_of_Independent_States')
plt.show()
```
